Conexão com o Banco de Dados

```python
import pandas as pd
from pathlib import Path
import sys

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from data.processed.consultas import get_connection

# Manter conexão aberta globalmente
conn = get_connection()
cursor = conn.cursor()
print("✓ Conexão Estável!")
print("=" * 80)

# Função auxiliar para executar queries facilmente
def consultar(query):
    """Executa uma query e retorna como DataFrame pandas"""
    cursor.execute(query)
    columns = [description[0] for description in cursor.description]
    data = cursor.fetchall()
    return pd.DataFrame(data, columns=columns)

# Visualizar tabela de clientes
print("\n📊 TABELA: CLIENTS")
print("-" * 80)
df_clients = consultar("SELECT * FROM clients")
display(df_clients)

print("\n📊 TABELA: LOANS")
print("-" * 80)
df_loans = consultar("SELECT * FROM loans")
display(df_loans)

print("\n" + "=" * 80)
print("✓ Conexão mantida aberta! Use a função consultar() para fazer queries.")
print("Exemplo: consultar('SELECT * FROM clients WHERE id = 1')")
print("=" * 80)
```

## 🔍 Exemplos de Uso

Agora que a conexão está aberta, você pode usar a função `consultar()` em qualquer célula:

**Exemplos:**
- `consultar("SELECT * FROM clients LIMIT 5")` - Ver primeiros 5 registros
- `consultar("SELECT COUNT(*) as total FROM clients")` - Contar registros
- `consultar("SELECT * FROM loans WHERE status='approved'")` - Filtrar dados
